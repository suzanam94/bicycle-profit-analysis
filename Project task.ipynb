```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv(r'C:\Users\PC\Desktop\Projekat\adventure_works_purchase_orders.csv')
```

```python
data.head()
```

```python
#Checking for null values
data.isnull().sum()
```

```python
#Changing data type into datetime
data['OrderDate'] = pd.to_datetime(data['OrderDate'])
data['DueDate'] = pd.to_datetime(data['DueDate'])
data['ShipDate'] = pd.to_datetime(data['ShipDate'])
#Adding new columns for Month and Year
data['Month'] = data['OrderDate'].dt.month

data['Year'] = data['OrderDate'].dt.year
```

```python
data.head()
```

```python
data.head()
```

```python
#checking for duplicates
data.duplicated().sum()
```

```python
# Calculate Total Sales, Total Cost and Total Profit
data['Total_Sales'] = data['OrderQty']*(data['UnitPrice'] - data['UnitPrice'] * data['UnitPriceDiscount'])
data['Total_Cost'] = data['OrderQty'] * data['Unit_Cost'] + data['Unit_Freight_Cost']
data['Profit'] = data['Total_Sales']-data['Total_Cost']
```

```python
data.describe()
```

```python
data.head()
```

```python
data['ShipDate'].unique()
```

```python
data['DueDate'].unique()
```

```python
# Removing the columns 'DueDate' and 'ShipDate' as they are not relevant to the analysis
data.drop(['DueDate', 'ShipDate'], axis=1, inplace=True)
```

```python
data.info()
```

```python
#Sales over years
yearly_sales = data.groupby('Year')['Total_Sales'].sum().reset_index()

yearly_sales
```

```python
plt.figure(figsize=(10,6))
plt.plot(yearly_sales['Year'], yearly_sales['Total_Sales'], marker='o', color='b')
plt.title('Total sales over the years')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

Sales decline in 2008


```python
# Profit over the years
yearly_profit = data.groupby('Year')['Profit'].sum().reset_index()

yearly_profit
```

```python
plt.figure(figsize=(10,6))
plt.plot(yearly_profit['Year'], yearly_profit['Profit'], marker='o', color='b')
plt.title('Profit over the years')
plt.xlabel('Year')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

Profit decline in 2008.

```python
# Trends by region
region_profit = data.loc[
    (data['OrderDate'].dt.year >= 2007) & (data['OrderDate'].dt.year <= 2008)
].groupby('Sales_Region')['Profit'].sum()


region_profit = region_profit.sort_values(ascending=False)
plt.figure(figsize=(8, 4))
region_profit.plot(kind='bar')
plt.title('Profit by Regions (2007-2008)')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()
```

```python
# Remove data for Germany as it is not relevant to the current analysis - it doesn't have data before 2008.
data = data.loc[data['Sales_Region'] != 'Germany']
```

```python
data['Sales_Region'].unique()
```

```python
# Monthly profit comparison by regions for 2007 and 2008
monthly_profit_comparison = data.loc[
    (data['Year'] >= 2007) & (data['Year'] <= 2008)
].groupby(['Year', 'Month', 'Sales_Region'])['Profit'].sum().reset_index()


monthly_profit_comparison
```

```python
# Pivot data for comparing profit by the regions
pivot_profit_regions = monthly_profit_comparison.pivot_table(
    index=['Month'], 
    columns=['Year', 'Sales_Region'], 
    values='Profit', 
    aggfunc='sum'
)


pivot_profit_regions
```

```python
# Line graph for every region comparisson 2007 vs 2008
regions = monthly_profit_comparison['Sales_Region'].unique()
plt.figure(figsize=(14, 8))

for region in regions:
    if (2007, region) in pivot_profit_regions and (2008, region) in pivot_profit_regions:
        plt.plot(
            pivot_profit_regions.index, 
            pivot_profit_regions[2007, region], 
            label=f'{region} 2007', 
            marker='o'
        )
        plt.plot(
            pivot_profit_regions.index, 
            pivot_profit_regions[2008, region], 
            label=f'{region} 2008', 
            marker='o'
        )

plt.title('Profit 2007 vs 2008 by Region', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Profit')
plt.xticks(pivot_profit_regions.index, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year & Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Analisys by products

```python
#Profit by category

# Filter the data for the years 2007 and 2008
filtered_data = data[data['OrderDate'].dt.year.isin([2007, 2008])]

# Group the filtered data by Product_Category
category_profit = filtered_data.groupby('Product_Category').agg({
    'Total_Sales': 'sum',
    'Total_Cost': 'sum',
    'Profit': 'sum'
}).reset_index()

category_profit
```

```python
#Profit by category vizualization
plt.figure(figsize=(10, 6))

# Create a bar plot to visualize profit by product category
sns.barplot(x='Product_Category', y='Profit', data=category_profit)
# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)
# Add a title to the plot
plt.title('Profit by Product Categories')

plt.show()
```

```python
# Profit by Product_Sub_Category
sub_category_profit = filtered_data.groupby('Product_Sub_Category').agg({
    'Total_Sales': 'sum',
    'Total_Cost': 'sum',
    'Profit': 'sum'
}).reset_index()

sub_category_profit
```

```python
# Profit sub category plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Profit', y='Product_Sub_Category', data=sub_category_profit, palette='viridis')

# Title and labels
plt.title('Profit by Product_Sub_Category', fontsize=14)
plt.xlabel('Profit', fontsize=12)
plt.ylabel('Product_Sub_Category', fontsize=12)


plt.tight_layout()
plt.show()
```

```python
#Checking for negative values in profit for Product_Sub_Category
negative_profit_subcategories = sub_category_profit[sub_category_profit['Profit'] < 0]
negative_profit_subcategories
```

Some of products have negative profit !

COST ANALISYS


```python
# Product Category comparison 2007 and 2008 by average unit cost
# Filter the data for the years 2007 and 2008
filtered_data = data[data['OrderDate'].dt.year.isin([2007, 2008])]
product_cost_comparison = filtered_data.groupby(['Product_Name', 'Year', 'Month'])['Unit_Cost'].mean().reset_index()

product_cost_comparison.head()
```

```python
# Average cost by product category 

avg_cost_by_category = filtered_data.groupby('Product_Category')['Unit_Cost'].mean().reset_index()

avg_cost_by_category.head()
```

```python
#Cost by the region and province
cost_by_region = filtered_data.groupby('Sales_Region')['Total_Cost'].sum().reset_index()

cost_by_province = filtered_data.groupby('Sales_Province')['Total_Cost'].sum().reset_index()

print(cost_by_region.head())
print(cost_by_province.head())
```

```python
# Cost by month and year
cost_by_month_year = filtered_data.groupby(['Year', 'Month'])['Total_Cost'].sum().reset_index()

print(cost_by_month_year.head())
```

```python
# Line plot for Cost by Month and Year
plt.figure(figsize=(12, 6))

sns.lineplot(data=cost_by_month_year, x='Month', y='Total_Cost', hue='Year', marker='o')

plt.title('Total Cost by Month and Year', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Cost', fontsize=12)
plt.xticks(ticks=range(1, 13), labels=[
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
])
plt.legend(title='Year', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

Costs in 2008 are higher than in 2007


```python
# Average price and cost of Unit over the years
avg_unit_cost_price = filtered_data.groupby(['Product_Sub_Category','Year']).agg({ 
    'Unit_Cost': 'mean',
    'UnitPrice': 'mean'
}).reset_index()

avg_unit_cost_price
```

```python
#Graph for Unit Price in 2007 and 2008

# Filter the data for 2007 and 2008
filtered_data_2007_2008 = avg_unit_cost_price[avg_unit_cost_price['Year'].isin([2007, 2008])]

# Create a bar plot for UnitPrice comparison
plt.figure(figsize=(14, 8))

sns.barplot(
    data=filtered_data_2007_2008,
    x='Product_Sub_Category',
    y='UnitPrice',
    hue='Year',
    palette='Set2'
)

# Customize the plot
plt.title('Comparison of Average Unit Price by Product Sub Category (2007 vs 2008)', fontsize=16)
plt.xlabel('Product Sub Category', fontsize=12)
plt.ylabel('Average Unit Price', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Year', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()

```

```python

# Filter the data for 2007 and 2008
filtered_data_2007_2008 = avg_unit_cost_price[avg_unit_cost_price['Year'].isin([2007, 2008])]

# Create a bar plot for Unit Cost comparison
plt.figure(figsize=(14, 8))

sns.barplot(
    data=filtered_data_2007_2008,
    x='Product_Sub_Category',
    y='Unit_Cost',
    hue='Year',
    palette='coolwarm'
)

# Customize the plot
plt.title('Comparison of Average Unit Cost by Product Sub Category (2007 vs 2008)', fontsize=16)
plt.xlabel('Product Sub Category', fontsize=12)
plt.ylabel('Average Unit Cost', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Year', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

Prices are not significantly changed from 2007, but costs are higher in 2008
